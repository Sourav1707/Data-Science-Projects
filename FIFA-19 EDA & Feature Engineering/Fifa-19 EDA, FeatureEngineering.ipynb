# Data Preprocessing

```python
#dropping coloums that are not useful for analysis
fifa01 = fifa.drop(['Photo','Flag','Club Logo'],axis=1)
fifa01
```

```python
## Value and wage coloumn conversion function

def fun_convert(df_value):
    try:
        value = float(df_value[1:-1])
        suffix = df_value[-1]
        
        if suffix == 'M':
            value = value * 1000000
        elif suffix == 'K':
            value = value * 1000
    except ValueError:
        value = 0
    return value
fifa01['Value']=fifa01['Value'].apply(fun_convert)
fifa01['Wage'] = fifa01['Wage'].apply(fun_convert)
fifa01
```

## Joined date Conversion

```python
fifa01['Joined']
```

```python
def date_conversion(date):
    if (pd.isna(date))!=True:
        val = []
        val = date.split(',')
        dat = int(val[1])
        return dat
    else:
        return date
```

```python
fifa01['Joined']=fifa01['Joined'].apply(date_conversion)
```

```python
fifa01['Joined']
```

## Height Conversion

```python
fifa01['Height']
```

```python
#in centimeter
def height_conversion(height):
    if (pd.isna(height)) != True:
        chk = str(height)
        h = []
        h = height.split("'")
        ft = float(h[0])
        if(h[1] != ''):
            inch = float(h[1])
        else:
            inch = 0
        tot_inch = inch + ft*12
        h = tot_inch * 2.54
        return h
    else:
        return height
```

```python
fifa01['Height'] = fifa01['Height'].apply(height_conversion)
```

```python
fifa01['Height'].isnull().sum()
```

```python
#replacing null values with mean
fifa01['Height'] = fifa01['Height'].fillna(value = fifa01['Height'].mean())
```

```python
fifa01['Height']
```

## Weight column conversion

```python
fifa01['Weight']
```

```python
#converting to kg

def weight_conversion(weight):
    if (pd.isna(weight)) != True:
        w = int(weight[0:-3])*0.453592
        return w
    else:
        return weight
```

```python
fifa01['Weight'] = fifa01['Weight'].apply(weight_conversion)
```

```python
fifa01['Weight'].isnull().sum()
```

```python
#replacing null values with mean
fifa01['Weight'] = fifa01['Weight'].fillna(value = fifa01['Weight'].mean())
```

```python
fifa01['Weight']
```

## Release Clause Conversion

```python
fifa01['Release Clause']
```

```python
def fun_convert(df_value):
    
    if(pd.isna(df_value))!= True:
        try:
            value = float(df_value[1:-1])
            suffix = df_value[-1]

            if suffix == 'M':
                value = value * 1000000
            elif suffix == 'K':
                value = value * 1000
        except ValueError:
            value = 0
        return value
    else:
        return df_value
```

```python
fifa01['Release Clause']=fifa01['Release Clause'].apply(fun_convert)
```
